```python
#24MAI0114
import category_encoders as ce
encoder = ce.BinaryEncoder(cols=['diagnosis'])
df = encoder.fit_transform(df)
```

```python
#24MAI0114
#check percentage of frequency distribution of diagnosi
percentage_distribution = df['diagnosis_0'].value_counts() / float(len(df))
print(percentage_distribution)
```

```python
X = df.drop(['diagnosis_0'], axis=1)

y = df['diagnosis_0']
```

```python
#24MAI0114
# Split data into separate training and test set 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
#24MAI0114
#Feature Scaling

cols = X_train.columns
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
#24MAI0114
X_train.head()
```

```python
#24MAI0114
#Fit K Neighbours Classifier to the training
# import KNeighbors ClaSSifier from sklearn
from sklearn.neighbors import KNeighborsClassifier


# instantiate the model
knn = KNeighborsClassifier(n_neighbors=3)


# fit the model to the training set
knn.fit(X_train, y_train)
```

```python
#24MAI0114
y_pred = knn.predict(X_test)

y_pred
```

```python
knn.predict_proba(X_test)[:,0]
```

```python

#24MAI0114
# probability of getting output as 4 - malignant cancer

knn.predict_proba(X_test)[:,1]
```

```python
#24MAI0114
#Checking the accuracy score
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
y_pred_train = knn.predict(X_train)
```

```python
#24MAI0114
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

```python
#Check for overfitting and underfitting
# print the scores on training and test set

print('Training set score: {:.4f}'.format(knn.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(knn.score(X_test, y_test)))
```

```python
#24MAI0114
#Confusion matrix
# Print the Confusion Matrix with k =3 and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
#24MAI0114
# visualize confusion matrix with seaborn heatmap
import matplotlib.pyplot as  plt
import seaborn as sns
plt.figure(figsize=(6,4))

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
plt.title("24MAI0114 ABHAYNATH K")
plt.show()
```
